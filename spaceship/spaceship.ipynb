```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sp
```

```python
x=np.ones(500)
```

```python
#a, b = sp.ellip(3, 2, 10, [.1,.9], 'bandpass', analog=False)
a1, b1 = sp.butter(1, [.01,.02], 'bandpass', analog=False)
a2, b2 = sp.butter(1, [.4,.5], 'bandpass', analog=False)
a3, b3 = sp.butter(1, [.89,.9], 'bandpass', analog=False)

A=[a1,a2,a3]
B=[b1,b2,b3]
```

```python
fig, axis=plt.subplots(1,2, figsize=(20,7))
t=np.arange(0,len(x))/60

for a,b in zip(A,B):
    W, H=sp.freqz(a,b, fs=1)
    axis[0].plot(W, np.abs(H))
    axis[0].set_title('Resposta em frequência')
    axis[0].set_xlabel('frequência normalizada')
    
    y=sp.lfilter(a,b,x)
    axis[1].plot(t,y)
    plt.title('Resposta ao degrau')
    axis[1].set_xlabel('tempo(s)')
    
plt.savefig('plot.png')
```
